```python
x = np.linspace(0,10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1,'k:',linewidth=3)
plt.plot(x,y2,'r--',linewidth=3)
plt.grid()
plt.ylabel('y', fontsize=14, color='black', fontstyle='italic')
plt.xlabel('x', fontsize=14, color='black', fontstyle='oblique')
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.legend(['sin(x)', 'cos(x)'])
plt.savefig('myFigure.png')
```
